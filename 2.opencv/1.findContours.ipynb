```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = r'./image_file/findContours/fire.png'
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # BGR
plt.imshow(gray, cmap='gray')
```

### 

```python
ret, binary = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY) # ret: binary
```

### findContours
### drawContours

```python
kernel = np.ones((1, 5), np.uint8)
binary = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, anchor=(2, 0), iterations=5)
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
cv2.drawContours(img, contours, -1, (0, 0, 255), 3)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### 

```python
for i in contours:
    area = cv2.contourArea(i)
    print(area)
```
